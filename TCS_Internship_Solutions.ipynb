# Clean the Data

# Handle missing values and ensure data integrity.

```python
# Drop rows with missing values
df = df.dropna(subset=["CustomerID", "Description"])

df
```

```python
# Remove rows with negative or zero quantities
df = df[df["Quantity"] > 0]

df
```

```python
# Explicitly create a copy of the DataFrame
df = df[df["Quantity"] > 0].copy()

# Add a new 'Revenue' column for task 1
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
# Adding a new column with a constant value
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df
```

# Analyze Key Revenue Drivers

Aggregate revenue by product and category.

```python
# Top products by revenue
product_revenue = df.groupby("Description")["Revenue"].sum().sort_values(ascending=False)

# Display the top 10 products
print(product_revenue.head(10))
```

```python
data = {"Product Revenue": product_revenue}
df_ProductRevenue = pd.DataFrame(data)
```

```python
df_ProductRevenue
```

# Geographic and Demographic Contributions
Aggregate revenue by country.

```python
# Revenue by country
country_revenue = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False)

# Display revenue by country
data = {"Country Revenue": country_revenue}
df_CountryRevenue = pd.DataFrame(data)


df_CountryRevenue
```

# Customer Purchase Patterns
Analyze customer behavior and retention.

```python
# Total revenue by customer
customer_revenue = df.groupby("CustomerID")["Revenue"].sum().sort_values(ascending=False)

# Frequency of purchases by customer
customer_frequency = df.groupby("CustomerID")["InvoiceNo"].nunique()

# Combine both metrics
customer_analysis = pd.DataFrame({
    "Revenue": customer_revenue,
    "Frequency": customer_frequency
})

data = {"Revenue": customer_revenue,"Frequency": customer_frequency}

df_Customer_Analysis = pd.DataFrame(data)

df_Customer_Analysis
```

# Seasonal Trends
Analyze revenue trends over time.

```python
# Convert InvoiceDate to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# Extract month and year for grouping
df["Month"] = df["InvoiceDate"].dt.to_period("M")

# Revenue by month
monthly_revenue = df.groupby("Month")["Revenue"].sum()

# Display monthly revenue
print(monthly_revenue)
```

# Operational Inefficiencies
Analyze returns and other operational issues.

```python
# Returns analysis (negative quantities)
returns = df[df["Quantity"] < 0]

# Total returns by product
returns_analysis = returns.groupby("Description")["Quantity"].sum().sort_values()

# Display the products with the highest returns
print(returns_analysis.head(10))
```

# Bar Chart for Top Products by Revenue

```python
import seaborn as sns

# Get top 10 products by revenue
top_products = product_revenue.sort_values(by="Revenue", ascending=False).head(10)

# Plot Revenue
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x="Description", y="Revenue", palette="viridis")
plt.title("Top 10 Products by Revenue", fontsize=16)
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Product", fontsize=12)
# Rotate x-axis labels
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Bar Chart for Top Countries by Revenue

```python
# Ensure country_revenue is a DataFrame (not a Series)
country_revenue = df.groupby("Country")["Revenue"].sum().reset_index()

# Get top 10 countries by revenue
top_countries = country_revenue.sort_values("Revenue", ascending=False).head(10)

# Plot  Revenue
plt.figure(figsize=(10, 6))
sns.barplot(data=top_countries, x="Country", y="Revenue", palette="coolwarm")
plt.title("Top 10 Countries by Revenue", fontsize=16)
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Country", fontsize=12)
# Rotate x-axis labels
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Segmenting Customers by Country and Aggregating Key Metrics

```python
# Aggregate revenue and purchase frequency by Country and CustomerID
demographic_analysis = df.groupby(["Country", "CustomerID"]).agg({
    "Revenue": "sum",          # Total revenue
    "InvoiceNo": "nunique",    # Purchase frequency
    "Quantity": "sum"          # Total quantity purchased
}).reset_index()

# Rename columns for better readability
demographic_analysis.rename(columns={"InvoiceNo": "PurchaseFrequency"}, inplace=True)

# Store the result in a DataFrame (already done with `demographic_analysis`)
df_demographic_analysis = demographic_analysis

df_demographic_analysis
```

```python
df.columns
```

# Visualize Customer Segmentation by Demographics
Total Revenue by Country

```python
# Revenue by Country (Demographic Analysis)
country_revenue = demographic_analysis.groupby("Country")["Revenue"].sum().reset_index()

# Sort by Revenue
country_revenue = country_revenue.sort_values("Revenue", ascending=False)

# Plot Revenue
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(data=country_revenue, x="Country", y="Revenue", palette="coolwarm")
plt.title("Total Revenue by Country", fontsize=16)
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Country", fontsize=12)
# Rotate x-axis labels
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Segmenting by Purchase Frequency
Analyze customers with high purchase frequency within each country.

```python
# Customers with high purchase frequency (top 5 per country)
top_customers_by_country = demographic_analysis.groupby("Country").apply(
    lambda x: x.nlargest(5, "PurchaseFrequency")
).reset_index(drop=True)

# Create a new DataFrame for the top customers
df_top_customers_by_country = top_customers_by_country

df_top_customers_by_country
```
